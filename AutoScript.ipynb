```python
%%writefile run_pipeline.py
import datetime
import os
import subprocess

os.makedirs("logs", exist_ok=True)

timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
log_file = f"logs/log_{timestamp}.txt"

with open(log_file, "w") as log:
    log.write("Pipeline started\n")

    result = subprocess.run(
        "python spect_automation.py",
        shell=True,
        capture_output=True,
        text=True
    )

    log.write("STDOUT:\n" + result.stdout + "\n")
    log.write("STDERR:\n" + result.stderr + "\n")

    log.write(f"Return code: {result.returncode}\n")

    if result.returncode != 0:
        log.write("Pipeline FAILED\n")
    else:
        log.write("Pipeline SUCCESS\n")

    log.write("Pipeline finished\n")

print(f"Log saved to {log_file}")
```
